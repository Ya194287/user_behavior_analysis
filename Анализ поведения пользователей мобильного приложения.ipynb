## Цель исследования

Цель проекта — проанализировать поведение пользователей мобильного приложения "Ненужные вещи", выделить группы пользователей на основе их действий и метрик, а также проверить гипотезы о различиях в поведении. Результаты исследования помогут понять, как пользователи взаимодействуют с приложением, и предложить рекомендации для повышения удержания и конверсии.

### Основные задачи:

1. Выделить группы пользователей, различающихся по метрикам:
   - Retention rate (удержание).
   - Время, проведенное в приложении.
   - Частота действий.
   - Конверсия в целевое действие (`contacts_show` — просмотр контактов).
2. Провести исследовательский анализ данных (EDA) для изучения поведения пользователей.
3. Сегментировать пользователей на основе их действий.
4. Проверить статистические гипотезы:
   - Гипотеза 1: Пользователи, установившие приложение из `yandex` и `google`, имеют разную конверсию в `contacts_show`.
   - Гипотеза 2 (собственная): Пользователи, добавляющие объявления в избранное (`favorites_add`), чаще совершают `contacts_show`.
5. Подготовить презентацию с выводами и рекомендациями в формате PDF.

```python
# Расчет метрик для всех пользователей

# Retention rate на 7 дней
first_event = df.groupby('user_id')['event_time'].min().reset_index()
first_event.columns = ['user_id', 'first_event_time']
df_ret = df.merge(first_event, on='user_id')
df_ret['days_since_first'] = (df_ret['event_time'] - df_ret['first_event_time']).dt.days
retention_7 = df_ret[(df_ret['days_since_first'] > 0) & (df_ret['days_since_first'] <= 7)]['user_id'].nunique() / df['user_id'].nunique()
print(f"\nRetention rate на 7 дней: {retention_7:.2%}")
```

Примерно четверть пользователей возвращается в приложение в течение недели после первого взаимодействия. Это умеренный показатель: он указывает на наличие заинтересованной аудитории, но также на потенциал для улучшения удержания.

```python
# Построение тепловой карты Retention Rate для всех пользователей

# Преобразуем все временные столбцы в datetime
df['event_time'] = pd.to_datetime(df['event_time'])
first_event['first_event_time'] = pd.to_datetime(first_event['first_event_time'])

# Для первого события сохраняем только дату (без времени) как datetime (не как date!)
first_event['first_event_date'] = first_event['first_event_time'].dt.normalize()  # Альтернатива: .dt.floor('D')

# Объединяем с основным датасетом
df_retention = df.merge(first_event[['user_id', 'first_event_date']], on='user_id')

# Нормализуем время события (оставляем только дату)
df_retention['event_date'] = df_retention['event_time'].dt.normalize()

# Вычисляем разницу в днях между событием и первым событием (теперь оба столбца в datetime)
df_retention['days_since_first'] = (df_retention['event_date'] - df_retention['first_event_date']).dt.days

# Ограничиваем анализ первыми 7 днями
df_retention = df_retention[df_retention['days_since_first'].between(0, 7)]

# Создаем когорты по дате первого события
cohort_data = df_retention.groupby(['first_event_date', 'days_since_first'])['user_id'].nunique().reset_index()

# Преобразуем данные в таблицу для тепловой карты
cohort_pivot = cohort_data.pivot(index='first_event_date', columns='days_since_first', values='user_id')

# Нормализуем по числу пользователей в когорте (на день 0)
cohort_sizes = cohort_pivot[0]  # Число пользователей в день 0 для каждой когорты
cohort_pivot_normalized = cohort_pivot.divide(cohort_sizes, axis=0)

# Строим тепловую карту
plt.figure(figsize=(12, 8))
sns.heatmap(
    cohort_pivot_normalized, 
    annot=True, 
    fmt='.1%',  # Формат процентов с одним знаком после запятой
    cmap='Blues', 
    cbar_kws={'label': 'Retention Rate'},
    vmin=0,  # Минимальное значение для цветовой шкалы
    vmax=1   # Максимальное значение
)
plt.title('Retention Rate по всем пользователям (первые 7 дней)')
plt.xlabel('Дни с первого события')
plt.ylabel('Дата первого события')
plt.tight_layout()  # Автоматическая подгонка layout
plt.show()
```

Наблюдается резкое падение Retention после 1-го дня, затем снижение замедляется, но к 7-му дню остается лишь небольшая доля пользователей.

Быстрое падение Retention после 1-го дня указывает на низкую вовлеченность новых пользователей, что требует улучшения онбординга и мотивации к возвращению.

```python
# Время в приложении (сессии: разрыв > 30 минут = новая сессия)
df_sorted = df.sort_values(['user_id', 'event_time'])
df_sorted['time_diff'] = df_sorted.groupby('user_id')['event_time'].diff()
session_threshold = timedelta(minutes=30)
df_sorted['new_session'] = (df_sorted['time_diff'] > session_threshold).astype(int).cumsum()
session_times = df_sorted.groupby(['user_id', 'new_session'])['event_time'].agg(['min', 'max'])
session_times['duration'] = (session_times['max'] - session_times['min']).dt.total_seconds() / 60  # в минутах
avg_session_time = session_times['duration'].mean()
print(f"\nСреднее время сессии: {avg_session_time:.2f} минут")
```

Для приложений вроде досок объявлений это нормальный показатель. Пользователи обычно заходят, чтобы быстро найти что-то нужное, а не "залипать" надолго, как в соцсетях или играх.

```python
# Сортируем данные по user_id и времени события
df_sorted = df.sort_values(['user_id', 'event_time'])

# Вычисляем разницу во времени между событиями для каждого пользователя
df_sorted['time_diff'] = df_sorted.groupby('user_id')['event_time'].diff()

# Определяем новую сессию (разрыв > 30 минут)
session_threshold = timedelta(minutes=30)
df_sorted['new_session'] = (df_sorted['time_diff'] > session_threshold).astype(int).cumsum()

# Группируем по пользователю и сессии, вычисляем начало и конец сессии
session_times = df_sorted.groupby(['user_id', 'new_session'])['event_time'].agg(['min', 'max'])

# Вычисляем длительность сессии в минутах
session_times['duration'] = (session_times['max'] - session_times['min']).dt.total_seconds() / 60  # в минутах

# Среднее время сессии
avg_session_time = session_times['duration'].mean()
print(f"\nСреднее время сессии: {avg_session_time:.2f} минут")

# Анализ распределения длительности сессий
print("\nОписательная статистика длительности сессий (в минутах):")
print(session_times['duration'].describe())

# Построим гистограмму распределения длительности сессий
plt.figure(figsize=(10, 5))
sns.histplot(session_times['duration'], bins=50, kde=True)
plt.title('Распределение длительности сессий (в минутах)')
plt.xlabel('Длительность сессии (минуты)')
plt.ylabel('Частота')
plt.show()

# Проверка аномально долгих сессий
# Определим аномально долгие сессии как те, что превышают 99-й процентиль
long_session_threshold = session_times['duration'].quantile(0.99)
long_sessions = session_times[session_times['duration'] > long_session_threshold]
print(f"\nАномально долгие сессии (длительность > {long_session_threshold:.2f} минут):")
print(f"Количество аномально долгих сессий: {len(long_sessions)}")
print("Примеры аномально долгих сессий:")
display(long_sessions.head())

# Проверка нулевых сессий (длительность = 0)
zero_sessions = session_times[session_times['duration'] == 0]
print(f"\nНулевые сессии (длительность = 0 минут):")
print(f"Количество нулевых сессий: {len(zero_sessions)}")

# Анализируем события в нулевых сессиях
if len(zero_sessions) > 0:
    zero_session_ids = zero_sessions.index.get_level_values('new_session')
    zero_session_users = zero_sessions.index.get_level_values('user_id')
    zero_session_events = df_sorted[(df_sorted['user_id'].isin(zero_session_users)) & 
                                   (df_sorted['new_session'].isin(zero_session_ids))]
    print("\nСобытия в нулевых сессиях:")
    print(zero_session_events['event_name'].value_counts())
    
    # Построим график распределения событий в нулевых сессиях
    plt.figure(figsize=(10, 5))
    zero_session_events['event_name'].value_counts().plot(kind='bar', color='lightcoral')
    plt.title('События в нулевых сессиях')
    plt.xlabel('Событие')
    plt.ylabel('Количество')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Нулевых сессий не обнаружено.")
```

```python
# 4.1 Подготовка данных для кластеризации

# Частота действий на пользователя
actions_per_user = df.groupby('user_id')['event_name'].count().reset_index()
actions_per_user.columns = ['user_id', 'actions']

# Конверсия в contacts_show (1 — совершил, 0 — нет)
conversion_per_user = df[df['event_name'] == 'contacts_show'].groupby('user_id')['event_name'].count().reset_index()
conversion_per_user['conversion'] = 1
conversion_per_user = conversion_per_user[['user_id', 'conversion']]
all_users = pd.DataFrame({'user_id': df['user_id'].unique()})
conversion_per_user = all_users.merge(conversion_per_user, on='user_id', how='left').fillna(0)
```

```python
# Retention (активность в течение 7 дней после первого события)
first_event = df.groupby('user_id')['event_time'].min().reset_index().rename(columns={'event_time': 'first_event_time'})
df_ret = df.merge(first_event, on='user_id')
df_ret['days_since_first'] = (df_ret['event_time'] - df_ret['first_event_time']).dt.days
retention_per_user = df_ret[(df_ret['days_since_first'] > 0) & (df_ret['days_since_first'] <= 7)].groupby('user_id')['event_name'].count().reset_index()
retention_per_user['retention'] = 1
retention_per_user = retention_per_user[['user_id', 'retention']]
retention_per_user = all_users.merge(retention_per_user, on='user_id', how='left').fillna(0)

# Объединяем метрики в один датафрейм
user_metrics = actions_per_user.merge(conversion_per_user, on='user_id').merge(retention_per_user, on='user_id')
print("Метрики для кластеризации (первые 5 строк):")
display(user_metrics.head())
```

```python
# 4.2 Кластеризация (K-means группы)

# Подготовка данных для кластеризации
X = user_metrics[['actions', 'conversion', 'retention']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Определение оптимального числа кластеров (метод локтя)
inertia = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 6), inertia, marker='o')
plt.title('Метод локтя для выбора числа кластеров')
plt.xlabel('Число кластеров')
plt.ylabel('Инерция')
plt.show()
```

3 кластера выглядят как оптимальный выбор, так как после этого добавление новых кластеров (4, 5) дает меньший прирост качества.

```python
# На основе графика
n_clusters = 3

# Кластеризация с выбранным числом кластеров
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
user_metrics['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Добавим тепловые карты по каждому кластеру

# 1. Преобразуем все временные столбцы в datetime
df['event_time'] = pd.to_datetime(df['event_time'])
df_with_clusters = df.merge(user_metrics[['user_id', 'cluster']], on='user_id')

# 2. Определяем дату первого события для каждого пользователя (сохраняем как datetime)
first_event = df_with_clusters.groupby('user_id')['event_time'].min().reset_index()
first_event.columns = ['user_id', 'first_event_datetime']
first_event['first_event_date'] = first_event['first_event_datetime'].dt.normalize()  # Оставляем только дату

# 3. Объединяем с датасетом
df_retention = df_with_clusters.merge(first_event[['user_id', 'first_event_date']], on='user_id')

# 4. Нормализуем время события (оставляем только дату)
df_retention['event_date'] = df_retention['event_time'].dt.normalize()

# 5. Вычисляем разницу в днях (теперь оба столбца в datetime)
df_retention['days_since_first'] = (df_retention['event_date'] - df_retention['first_event_date']).dt.days

# 6. Ограничиваем анализ первыми 7 днями
df_retention = df_retention[df_retention['days_since_first'].between(0, 7)]

# 7. Строим тепловые карты для каждого кластера
for cluster in sorted(user_metrics['cluster'].unique()):
    # Фильтруем данные по кластеру
    cluster_data = df_retention[df_retention['cluster'] == cluster]
    
    # Создаем когорты по дате первого события
    cohort_data = cluster_data.groupby(['first_event_date', 'days_since_first'])['user_id'].nunique().reset_index()
    
    # Преобразуем в таблицу для тепловой карты
    try:
        cohort_pivot = cohort_data.pivot(index='first_event_date', columns='days_since_first', values='user_id')
        
        # Пропускаем кластеры с недостаточным количеством данных
        if len(cohort_pivot) < 2 or 0 not in cohort_pivot.columns:
            print(f"Недостаточно данных для кластера {cluster}, пропускаем")
            continue
            
        # Нормализуем по числу пользователей в когорте (на день 0)
        cohort_sizes = cohort_pivot[0]
        cohort_pivot_normalized = cohort_pivot.divide(cohort_sizes, axis=0)
        
        # Строим тепловую карту
        plt.figure(figsize=(10, 6))
        sns.heatmap(
            cohort_pivot_normalized, 
            annot=True, 
            fmt='.1%',
            cmap='Blues', 
            cbar_kws={'label': 'Retention Rate'},
            vmin=0,
            vmax=1
        )
        plt.title(f'Retention Rate для кластера {cluster}\n(Размер когорты: {cohort_sizes.mean():.0f} пользователей)')
        plt.xlabel('Дни с первого события')
        plt.ylabel('Дата первого события')
        plt.tight_layout()
        plt.show()
        
    except Exception as e:
        print(f"Ошибка при обработке кластера {cluster}: {str(e)}")
        continue
```

День 1: Retention варьируется от 1.18% до 8.96% (например, 5.15% для когорты 2019-10-07).

5. **Проверка гипотез**
   - **Гипотеза 1:** Пользователи, установившие приложение из `yandex` и `google`, демонстрируют разную конверсию в `contacts_show`.
     - H0: Конверсия в `contacts_show` одинакова для пользователей из `yandex` и `google`.
     - H1: Конверсия в `contacts_show` различается между пользователями из `yandex` и `google`.

```python
# Подготовка данных
yandex_users = df[df['source'] == 'yandex']['user_id'].nunique()
google_users = df[df['source'] == 'google']['user_id'].nunique()

yandex_conversions = df[(df['source'] == 'yandex') & (df['event_name'] == 'contacts_show')]['user_id'].nunique()
google_conversions = df[(df['source'] == 'google') & (df['event_name'] == 'contacts_show')]['user_id'].nunique()

# Конверсии
yandex_conversion_rate = yandex_conversions / yandex_users
google_conversion_rate = google_conversions / google_users

print("Конверсия по источникам:")
print(f"yandex: {yandex_conversion_rate:.2%} ({yandex_conversions}/{yandex_users})")
print(f"google: {google_conversion_rate:.2%} ({google_conversions}/{google_users})")
```

```python
# Z-тест для пропорций
count = np.array([yandex_conversions, google_conversions])
nobs = np.array([yandex_users, google_users])
stat, p_value = proportions_ztest(count, nobs)

print("\nZ-тест для гипотезы 1:")
print(f"Z-статистика: {stat:.2f}")
print(f"p-value: {p_value:.4f}")
if p_value < 0.05:
    print("Отвергаем нулевую гипотезу: конверсия между yandex и google различается (p < 0.05).")
else:
    print("Не отвергаем нулевую гипотезу: нет значимых различий в конверсии между yandex и google (p >= 0.05).")
```

   - **Гипотеза 2:** Пользователи, которые добавляют объявления в избранное (`favorites_add`), демонстрируют более высокую конверсию в просмотр контактов (`contacts_show`), что указывает на потенциал упрощения доступа к контактам из раздела избранного для повышения общей конверсии.
     - H0: Добавление объявлений в избранное не влияет на конверсию в `contacts_show`; конверсия одинакова для пользователей с `favorites_add` и без него.
     - H1: Пользователи с `favorites_add` имеют более высокую конверсию в `contacts_show`, и улучшение доступа к контактам из избранного может увеличить конверсию.

```python
# Подготовка данных: создаем таблицу сопряженности
users_favorites = set(df[df['event_name'] == 'favorites_add']['user_id'])  # Пользователи, совершившие favorites_add
users_contacts = set(df[df['event_name'] == 'contacts_show']['user_id'])  # Пользователи, совершившие contacts_show
all_users = set(df['user_id'])  # Все пользователи в датасете

# Таблица сопряженности
# favorites_add: 1 — совершил, 0 — нет
# contacts_show: 1 — совершил, 0 — нет
contingency_table = pd.DataFrame({
    'favorites_add_yes': [
        len(users_favorites & users_contacts),  # favorites_add = 1, contacts_show = 1
        len(users_favorites - users_contacts)   # favorites_add = 1, contacts_show = 0
    ],
    'favorites_add_no': [
        len(users_contacts - users_favorites),  # favorites_add = 0, contacts_show = 1
        len(all_users - users_favorites - users_contacts)  # favorites_add = 0, contacts_show = 0
    ]
}, index=['contacts_show_yes', 'contacts_show_no'])

print("\nТаблица сопряженности (favorites_add vs contacts_show):")
display(contingency_table)

# Проверка: сумма всех значений должна равняться общему числу пользователей
total_users_in_table = contingency_table.values.sum()
total_users = len(all_users)
print(f"\nПроверка: общее число пользователей в таблице: {total_users_in_table}")
print(f"Общее число пользователей в датасете: {total_users}")
if total_users_in_table == total_users:
    print("Таблица сопряженности построена корректно: все пользователи учтены.")
else:
    print("Ошибка: число пользователей в таблице не совпадает с общим числом пользователей.")
```

Из пользователей, которые просмотрели контакты (contacts_show_yes), 13,9% (136 из 979) добавили товар/услугу в избранное.

Из тех, кто не смотрел контакты (contacts_show_no), только 6,5% (215 из 3314) добавили в избранное.

Влияние просмотра контактов:

Пользователи, открывшие контакты, в 2,1 раза чаще сохраняют товар в избранное.

Это говорит о том, что просмотр контактов — сильный индикатор заинтересованности.

Потенциальные улучшения:

Можно стимулировать пользователей чаще открывать контакты (например, улучшая доверие к карточке товара).

У тех, кто не смотрит контакты, низкая вовлечённость – возможно, стоит доработать UX или мотивацию.

```python
# Хи-квадрат тест
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nХи-квадрат тест для гипотезы 2:")
print(f"Хи-квадрат статистика: {chi2:.2f}")
print(f"p-value: {p:.4f}")
print(f"Степени свободы: {dof}")
if p < 0.05:
    print("Отвергаем нулевую гипотезу: есть связь между favorites_add и contacts_show (p < 0.05).")
else:
    print("Не отвергаем нулевую гипотезу: нет значимой связи между favorites_add и contacts_show (p >= 0.05).")
```

```python
# Дополнительно: конверсия в contacts_show в зависимости от favorites_add
conversion_with_favorites = contingency_table.loc['contacts_show_yes', 'favorites_add_yes'] / (contingency_table['favorites_add_yes'].sum())
conversion_without_favorites = contingency_table.loc['contacts_show_yes', 'favorites_add_no'] / (contingency_table['favorites_add_no'].sum())

print("\nКонверсия в contacts_show в зависимости от favorites_add:")
print(f"С favorites_add: {conversion_with_favorites:.2%}")
print(f"Без favorites_add: {conversion_without_favorites:.2%}")

# Визуализация
conversion_favorites_data = pd.DataFrame({
    'favorites_add': ['С favorites_add', 'Без favorites_add'],
    'conversion_rate': [conversion_with_favorites, conversion_without_favorites]
})
plt.figure(figsize=(8, 5))
sns.barplot(data=conversion_favorites_data, x='favorites_add', y='conversion_rate', palette='Set3')
plt.title('Конверсия в contacts_show в зависимости от favorites_add')
plt.xlabel('Действие favorites_add')
plt.ylabel('Конверсия')
plt.ylim(0, 1)
for i, rate in enumerate(conversion_favorites_data['conversion_rate']):
    plt.text(i, rate + 0.01, f'{rate:.2%}', ha='center')
plt.show()
```

Пользователи, которые добавляли товар в избранное, в 1.8 раза чаще просматривали контакты (38.75% против 21.44%).

Добавление в избранное (favorites_add) — ключевой индикатор заинтересованности. Такие пользователи активнее переходят к контактам.

Стимулирование добавления в избранное (например, через уведомления или бонусы) — это способ повысить конверсию в contacts_show.
